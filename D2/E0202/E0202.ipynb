```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
import seaborn as sns
```

# 0202a

```python
# Import data
dataset = pd.read_csv('boston_housing.csv')

# Check corr between vars
cm = dataset.corr().round(2)
plt.figure(figsize=(13, 13))
sns.heatmap(data=cm, annot=True, cmap='Blues')
```

```python
# Define X, Y
x = dataset.loc[:,("rm","ptratio","lstat")].values
y = dataset.loc[:,"medv"].values.reshape((-1,1))

# Split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=1/4)

# Scaling
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)

# Linear regression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
# Evaluate
coef=regressor.coef_
intercept = regressor.intercept_
mse = mean_squared_error(y_test,regressor.predict(x_test))
variance = np.var(regressor.predict(x_test))
r2 = r2_score(y_test, regressor.predict(x_test))

print("Coefficients:",coef)
print("Intercept:", intercept)
print("MSE test set:", mse)
print("Variance test set:", variance)
print("r2 test set:", r2)
```

```python
# Plot
y_pred = regressor.predict(x_test)
plt.scatter(y_test,y_pred)
plt.title('Actual vs. predicted prices')
plt.xlabel('Actual price')
plt.ylabel('Predicted price')
plt.show()
```
