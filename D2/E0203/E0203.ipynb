```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

# 0203a

```python
# Import data
dataset = pd.read_csv('Position_Salaries.csv')

# Visualize
plt.scatter(dataset['Level'], dataset['Salary'])
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

```python
# Select data
x = dataset['Level'].values.reshape((-1,1))
y = dataset['Salary'].values.reshape((-1,1))

# Fit linear regression
regressor_linear = LinearRegression().fit(x,y)
prediction_linear = regressor_linear.predict(x)

# Fit polynominal regression
polynomial = PolynomialFeatures(degree = 3)
x_polynomial = polynomial.fit_transform(x)
regressor_poly = LinearRegression().fit(x_polynomial,y)
prediction_poly = regressor_poly.predict(x_polynomial)

# Plot
plt.scatter(x,y, color='black')
plt.plot(x, prediction_linear, color='red', label='Linear')
plt.plot(x, prediction_poly, color='blue', label='Polynomial')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.legend()
plt.show()
```
